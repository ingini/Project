```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
cv = pd.read_excel('0717_편의점.xlsx', index_col=0)
```

```python
la = pd.read_excel('0717_랄라블라.xlsx', index_col=0)
```

```python
cv_col = cv.columns[13:]
cv_col
```

```python
la_col = la.columns[13:]
la_col
```

```python
for i in cv_col:
    m = 2
    temp = cv[i].rolling(m).mean()
    print('\n컬럼 : {}\n'.format(i))
    while m < 10:
        m += 1
        temp = cv[i].rolling(m).mean()
        print('이동평균, 첨도 값은 : {}, {}'.format(m, temp.kurtosis()))
```

```python
for i in la_col:
    m = 2
    temp = la[i].rolling(m).mean()
    print('\n컬럼 : {}\n'.format(i))
    while m < 50:
        m += 1
        temp = la[i].rolling(m).mean()
        print('이동평균, 첨도 값은 : {}, {}'.format(m, temp.kurtosis()))
```

```python
for i in cv_col:
    cv['{}_mean'.format(i)] = round(cv[i].rolling(window=7).mean(), 1)
```

```python
for i in la_col:
    la['{}_mean'.format(i)] = round(la[i].rolling(window=7).mean(), 1)
```

```python
plt.rcParams['font.family'] = 'NanumGothic'
plt.rcParams['font.size'] = 12
plt.style.use(['dark_background'])
```

```python
plt.figure(figsize=(20,60))
for i,j in enumerate(cv_col):
    plt.subplot(12,1,i+1)
    plt.title(j)
    plt.plot(cv[j], color='w', label=j)
    plt.plot(cv['{}_mean'.format(j)], color='r', label='{}_mean'.format(j))
    plt.legend()
```

```python
plt.figure(figsize=(20,50))
for i,j in enumerate(la_col):
    plt.subplot(10,1,i+1)
    plt.title(j)
    plt.plot(la[j], color='w', label=j)
    plt.plot(la['{}_mean'.format(j)], color='r', label='{}_mean'.format(j))
    plt.legend()
```
