### preprocessing

```python
df['year'] = (df['date']/10000).astype('int')
```

```python
df['month'] = (df['year']/100).astype('int')
```

```python
df = df.rename(columns = {'Unnamed: 0':'year'})
```

```python
age = df['age'].unique()
age_point_replace = {'00~19' : 1, '20~39' : 2, '40~59' : 3, '60~99' : 4}
df['age'].apply(lambda x : age_point_replace.get(x))
```

```python
df['age'] = df['age'].apply(lambda x : age_point_replace.get(x))
```
