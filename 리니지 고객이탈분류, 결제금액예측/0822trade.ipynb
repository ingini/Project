```python
# module import
import pandas as pd
import numpy as np
```

```python
combat = pd.read_csv('test2_trade.csv')
trade = pd.read_csv('train_trade.csv')
```

```python
combat.tail()
```

```python
# make new column
combat["wk"] = 0
```

```python
# month classification by week
def wk_classification(df):
    df.loc[df['day'] <= 7, ['wk']] = 1
    df.loc[(df['day'] > 7) & (df['day'] <= 14), ['wk']] = 2
    df.loc[(df['day'] > 14) & (df['day'] <= 21), ['wk']] = 3
    df.loc[df['day'] >  21, ['wk']] = 4
    return

wk_classification(combat)
```

```python
# set index for groupby
combat.set_index(combat["source_acc_id"], inplace = True)
```

```python
combat.columns[:8]
```

```python
# delete column to same index
combat = combat.drop(combat.columns[:8], axis=1) 
```

```python
# confirm indexing
combat.sort_index(ascending = True).head(10)
```

```python
# groupby 'acc_id', "wk" and sum of other columns
t_group = combat.groupby(['source_acc_id', "wk"], sort = ['source_acc_id'])[combat.columns[:]].agg('sum')
t_group.head()
```

```python
act_wk1 = combat[combat["wk"]==1]
act_wk1.sort_index(ascending = True).head()
```

```python
combat.columns
```

```python
# delete column to same index
combat = combat.drop(['item_type'], axis=1) 
```

```python
combat.fillna(0)
```

```python
act_wk1 = combat[combat["wk"]==1].groupby("source_acc_id").sum()
act_wk1.rename(columns={ 'item_amount':'item_amount_1','item_price':'item_price_1'},inplace=True)
act_wk2 = combat[combat["wk"]==2].groupby("source_acc_id").sum()
act_wk2.rename(columns={ 'item_amount':'item_amount_2','item_price':'item_price_2'},inplace=True)
act_wk3 = combat[combat["wk"]==3].groupby("source_acc_id").sum()
act_wk3.rename(columns={ 'item_amount':'item_amount_3','item_price':'item_price_3'},inplace=True)
act_wk4 = combat[combat["wk"]==4].groupby("source_acc_id").sum()
act_wk4.rename(columns={ 'item_amount':'item_amount_4','item_price':'item_price_4'},inplace=True)
print(act_wk3)
```

```python
act_wk1.reset_index(inplace=True)
act_wk2.reset_index(inplace=True)
act_wk3.reset_index(inplace=True)
act_wk4.reset_index(inplace=True)
```

```python
del act_wk1["wk"]
del act_wk2["wk"]
del act_wk3["wk"]
del act_wk4["wk"]
```

```python
print(act_wk1.shape)
print(act_wk2.shape)
print(act_wk3.shape)
print(act_wk4.shape)
```

```python
act_mer1 = pd.merge(act_wk1, act_wk2, how='outer').fillna(0)
act_mer1.head(10)
act_mer1.shape
```

```python
act_mer2 = pd.merge(act_wk3, act_wk4, how='outer').fillna(0)
act_mer2.head(10)
act_mer2.shape
```

```python
act_mer = pd.merge(act_mer1, act_mer2, how = "outer").fillna(0)
act_mer
```

```python
act_mer.shape
```

```python
act_mer.to_csv('test2_source_acc_id.csv',index=False)
```
