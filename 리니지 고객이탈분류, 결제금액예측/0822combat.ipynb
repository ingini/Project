```python
# make new column
combat["wk"] = 0
```

```python
# month classification by week
def wk_classification(df):
    df.loc[df['day'] <= 7, ['wk']] = 1
    df.loc[(df['day'] > 7) & (df['day'] <= 14), ['wk']] = 2
    df.loc[(df['day'] > 14) & (df['day'] <= 21), ['wk']] = 3
    df.loc[df['day'] >  21, ['wk']] = 4
    return

wk_classification(combat)
```

```python
ls = ['pledge_cnt', 'random_attacker_cnt', 'random_defender_cnt', 'temp_cnt',
       'same_pledge_cnt', 'etc_cnt', 'num_opponent', 'wk']
```

```python
act_wk1 = combat[combat["wk"]==1].groupby("acc_id").sum()
act_wk1.rename(columns={'pledge_cnt':'pledge_cnt_1', 'random_attacker_cnt':'random_attacker_cnt_1', 'random_defender_cnt':'random_defender_cnt_1', 'temp_cnt':'temp_cnt_1', 'same_pledge_cnt':'same_pledge_cnt_1', 'etc_cnt':'etc_cnt_1', 'num_opponent':'num_opponent_1'},inplace=True)
act_wk2 = combat[combat["wk"]==2].groupby("acc_id").sum()
act_wk2.rename(columns={'pledge_cnt':'pledge_cnt_2', 'random_attacker_cnt':'random_attacker_cnt_2', 'random_defender_cnt':'random_defender_cnt_2', 'temp_cnt':'temp_cnt_2', 'same_pledge_cnt':'same_pledge_cnt_2', 'etc_cnt':'etc_cnt_2', 'num_opponent':'num_opponent_2'},inplace=True)
act_wk3 = combat[combat["wk"]==3].groupby("acc_id").sum()
act_wk3.rename(columns={'pledge_cnt':'pledge_cnt_3', 'random_attacker_cnt':'random_attacker_cnt_3', 'random_defender_cnt':'random_defender_cnt_3', 'temp_cnt':'temp_cnt_3', 'same_pledge_cnt':'same_pledge_cnt_3', 'etc_cnt':'etc_cnt_3', 'num_opponent':'num_opponent_3'},inplace=True)
act_wk4 = combat[combat["wk"]==4].groupby("acc_id").sum()
act_wk4.rename(columns={'pledge_cnt':'pledge_cnt_4', 'random_attacker_cnt':'random_attacker_cnt_4', 'random_defender_cnt':'random_defender_cnt_4', 'temp_cnt':'temp_cnt_4', 'same_pledge_cnt':'same_pledge_cnt_4', 'etc_cnt':'etc_cnt_4', 'num_opponent':'num_opponent_4'},inplace=True)
print(act_wk1)
```

```python
act_wk1.reset_index(inplace=True)
act_wk2.reset_index(inplace=True)
act_wk3.reset_index(inplace=True)
act_wk4.reset_index(inplace=True)
```

```python
del act_wk1["wk"]
del act_wk2["wk"]
del act_wk3["wk"]
del act_wk4["wk"]
```

```python
act_mer1 = pd.merge(act_wk1, act_wk2, how='outer', on = "acc_id")
act_mer1.head(10)
```

```python
act_mer2 = pd.merge(act_wk3, act_wk4, how='outer', on = "acc_id")
act_mer2.head(10)
```

```python
act_mer = pd.merge(act_mer1, act_mer2, how = "outer", on = "acc_id").fillna(0)
act_mer
```
